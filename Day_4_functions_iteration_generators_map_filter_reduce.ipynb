
# Functions, Iterators, Generators, Map, Filter and Reduce


# Functions in Python

### What is a Function?
A function is a reusable block of code that performs a specific task.

### Why Use Functions?
- Avoid code duplication
- Improve readability
- Easier debugging
- Better code organization
- Reusability

### Real Life Example
ATM Machine:
Input → Processing → Output

Similarly:
Arguments → Function → Result


```python

def greet():
    print("Welcome to Python")

greet()
```

# Function Parameters and Arguments

### Parameters vs Arguments

Parameter:
Variable defined inside function definition.

Argument:
Actual value passed during function call.


```python

def add(a, b):
    print(a+b)

add(10,20)
```

# Default, Keyword and Positional Arguments

### Positional Arguments
Order matters.

### Keyword Arguments
Order does not matter.

### Default Arguments
Default value is used if no argument is passed.


```python

def employee(name, age=21, dept="IT"):
    print(name, age, dept)

employee("Yaswanth")
employee("Yaswanth",25)
employee(age=25,name="Yaswanth",dept="AI")
```

# Return Statement

### return vs print

print():
- Displays output

return:
- Sends value back
- Can be reused later
- Ends function execution


```python

def square(x):
    return x*x

result = square(5)
print(result)
print(result+10)
```

# Variable Length Arguments

### *args
Accepts multiple positional arguments.

### **kwargs
Accepts multiple keyword arguments.


```python

def total(*args):
    print(sum(args))

total(10,20,30,40)

def info(**kwargs):
    print(kwargs)

info(name="Yaswanth",city="Buffalo")
```

# Lambda Functions

### Lambda Functions

Anonymous one-line functions.

Syntax:
lambda arguments : expression


```python

square = lambda x: x*x
print(square(6))
```

# Iterators

### Iterators

Iterator Protocol:
- __iter__()
- __next__()

Functions:
- iter()
- next()


```python

nums=[10,20,30]

it=iter(nums)

print(next(it))
print(next(it))
print(next(it))
```

# StopIteration

When iterator elements finish,
Python raises StopIteration.


```python

nums=[1,2]

it=iter(nums)

try:
    while True:
        print(next(it))
except StopIteration:
    print("Iterator Finished")
```

# Generators

### Generators

Generators use yield.

Advantages:
- Memory efficient
- Faster for large datasets
- Generates values when needed


```python

def count():
    for i in range(1,6):
        yield i

for x in count():
    print(x)
```

# Generator vs Normal Function

Normal Function:
Returns entire data at once.

Generator:
Returns values one by one.


```python

def normal():
    return [1,2,3]

def gen():
    yield 1
    yield 2
    yield 3

print(normal())

for x in gen():
    print(x)
```

# Map Function

### map()

Transforms each element.

Syntax:
map(function, iterable)


```python

nums=[1,2,3,4,5]

result=map(lambda x:x*x, nums)

print(list(result))
```

# Map Real World Example

Adding 18% GST to prices.


```python

prices=[100,200,300]

gst=list(map(lambda x:x*1.18, prices))

print(gst)
```

# Filter Function

### filter()

Keeps only values that satisfy condition.

Syntax:
filter(function, iterable)


```python

nums=[1,2,3,4,5,6]

even=filter(lambda x:x%2==0, nums)

print(list(even))
```

# Filter Real World Example

Filter students who passed.


```python

marks=[35,72,88,29,45]

passed=list(filter(lambda x:x>=40, marks))

print(passed)
```

# Reduce Function

### reduce()

Combines all elements into one final value.

Need:
from functools import reduce


```python

from functools import reduce

nums=[1,2,3,4]

result=reduce(lambda x,y:x+y, nums)

print(result)
```

# Reduce Largest Number

Finding maximum element using reduce.


```python

from functools import reduce

nums=[10,50,20,70,30]

largest=reduce(lambda a,b:a if a>b else b, nums)

print(largest)
```

# Map vs Filter vs Reduce

| Function | Purpose |
|----------|----------|
| map() | Transform |
| filter() | Select |
| reduce() | Aggregate |

Remember:
Map -> Change Data
Filter -> Choose Data
Reduce -> Combine Data


```python

nums=[1,2,3,4,5]

mapped=list(map(lambda x:x*2, nums))
filtered=list(filter(lambda x:x%2==0, nums))

from functools import reduce
reduced=reduce(lambda x,y:x+y, nums)

print(mapped)
print(filtered)
print(reduced)
```
