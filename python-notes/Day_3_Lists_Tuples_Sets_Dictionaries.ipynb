## Lists

A list is a container used to store multiple values in a single variable.

Characteristics:
- Ordered
- Mutable
- Allows duplicate values
- Can store different data types
- Supports indexing and slicing
- Uses square brackets []

```python
courses = ['Python', 'SQL', 'Machine Learning', 'Deep Learning']
print(courses)
```

### Why Lists?

Lists are useful when we need to store a collection of related items and the data may change over time.

## Indexing

Indexing is used to access a single element from a list. Python uses zero-based indexing.

```python
courses = ['Python', 'SQL', 'Machine Learning', 'Deep Learning']
print(courses[0])
print(courses[2])
```

### Negative Indexing

Negative indexes start from the end of the list.

-1 refers to the last element.

```python
courses = ['Python', 'SQL', 'Machine Learning', 'Deep Learning']
print(courses[-1])
print(courses[-2])
```

## Slicing

Slicing extracts multiple elements from a list.

Syntax:
list[start:end]

Start is included and end is excluded.

```python
numbers = [10,20,30,40,50,60]
print(numbers[1:4])
print(numbers[2:])
print(numbers[::-1])
```

## Concatenating Lists

The + operator combines lists and creates a new list.

```python
core = ['Python','SQL']
ai = ['Machine Learning','Deep Learning']
print(core + ai)
```

## Repeating Lists

The * operator repeats the entire list multiple times.

```python
skills = ['Python','SQL']
print(skills * 3)
```

## append()

append() adds a single element at the end of a list.

The original list is modified.

```python
topics = ['Python','SQL']
topics.append('AI')
print(topics)
```

## extend()

extend() adds multiple elements individually into the list.

```python
list1 = [1,2,3]
list2 = [4,5]
list1.extend(list2)
print(list1)
```

### append() vs extend()

append() adds one object.

extend() adds each element separately.

```python
a=[1,2]
b=[3,4]

x=a.copy(); x.append(b)
print(x)

y=a.copy(); y.extend(b)
print(y)
```

## pop()

pop() removes and returns an element.

By default it removes the last element.

```python
numbers=[10,20,30]
removed=numbers.pop()
print(removed)
print(numbers)
```

## remove()

remove() deletes an element by value.

```python
numbers=[10,20,30,40]
numbers.remove(20)
print(numbers)
```

## insert()

insert() adds an element at a specific position.

```python
numbers=[10,20,40]
numbers.insert(2,30)
print(numbers)
```

## sort()

sort() arranges elements in ascending order permanently.

```python
scores=[88,65,92,70]
scores.sort()
print(scores)
```

## reverse()

reverse() reverses the list permanently.

```python
scores=[88,65,92,70]
scores.reverse()
print(scores)
```

## count()

count() returns the number of occurrences of a value.

```python
numbers=[1,2,2,3,2]
print(numbers.count(2))
```

## index()

index() returns the position of an element.

```python
numbers=[10,20,30]
print(numbers.index(20))
```

## Nested Lists

A list inside another list is called a nested list.

```python
students=[['Yaswanth',24],['Rahul',23],['Priya',22]]
print(students[0])
print(students[1][1])
```

## List Comprehension

A shorter way to create lists using loops.

```python
squares=[num**2 for num in range(1,6)]
print(squares)
```

## Tuples

A tuple stores multiple values in a single variable.

Major difference from lists: tuples are immutable.

```python
student=('Yaswanth','UB','MS CSE')
print(student)
```

### Tuple Indexing and Slicing

```python
student=('Yaswanth','UB','MS CSE')
print(student[0])
print(student[:2])
```

## Tuple Packing

Storing multiple values into a tuple.

```python
employee=('Yaswanth',24,'AI/ML')
```

## Tuple Unpacking

Extracting values from a tuple into separate variables.

```python
employee=('Yaswanth',24,'AI/ML')
name,age,track=employee
print(name)
print(age)
print(track)
```

## Single Value Tuple

A comma is required when creating a tuple with only one value.

```python
num1=(10)
num2=(10,)
print(type(num1))
print(type(num2))
```

## Tuple Methods

```python
numbers=(1,2,2,3,2)
print(numbers.count(2))
print(numbers.index(3))
```

## Sets

A set is an unordered collection of unique values.

Duplicate values are automatically removed.

```python
skills={'Python','SQL','Python','AI'}
print(skills)
```

## add()

```python
skills={'Python','SQL'}
skills.add('AI')
print(skills)
```

## remove() vs discard()

remove() raises an error if the value does not exist.

discard() does not raise an error.

```python
skills={'Python','SQL'}
skills.discard('Java')
print(skills)
```

## Union

Combines all unique values from both sets.

```python
A={1,2,3}
B={3,4,5}
print(A|B)
```

## Intersection

Returns common values only.

```python
A={1,2,3}
B={2,3,4}
print(A&B)
```

## Difference

Returns values present in the first set but not the second.

```python
A={1,2,3}
B={2,3,4}
print(A-B)
```

## Symmetric Difference

Returns values that are not common in both sets.

```python
A={1,2,3}
B={2,3,4}
print(A^B)
```

## issubset() and issuperset()

```python
A={1,2}
B={1,2,3,4}
print(A.issubset(B))
print(B.issuperset(A))
```

## Dictionaries

A dictionary stores data in key:value pairs.

Keys must be unique while values can repeat.

```python
student={'name':'Yaswanth','university':'UB','program':'MS CSE'}
print(student)
```

## Accessing Values

Values can be accessed using keys.

```python
student={'name':'Yaswanth','gpa':3.8}
print(student['name'])
```

## get()

get() is safer because it returns None instead of an error when a key is missing.

```python
student={'name':'Yaswanth'}
print(student.get('gpa'))
```

## Adding New Data

```python
student={'name':'Yaswanth'}
student['gpa']=3.8
print(student)
```

## Updating Existing Data

```python
student={'gpa':3.8}
student['gpa']=4.0
print(student)
```

## Removing Data with pop()

```python
student={'name':'Yaswanth','gpa':4.0}
student.pop('gpa')
print(student)
```

## keys(), values() and items()

```python
student={'name':'Yaswanth','gpa':4.0}
print(student.keys())
print(student.values())
print(student.items())
```

## Looping Through Dictionaries

We can loop through keys, values, or both.

```python
student={'name':'Yaswanth','gpa':4.0}
for key,value in student.items():
    print(key,value)
```

## When to Use What?

List:
- Ordered data
- Data changes frequently

Tuple:
- Fixed data

Set:
- Unique values only

Dictionary:
- Data has labels
