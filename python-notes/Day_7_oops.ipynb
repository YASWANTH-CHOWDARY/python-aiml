## Day 7: OOP and File I/O 

Object Oriented Programming (OOP) is a way to organize code by bundling related data and functions together into objects. Instead of having tons of separate variables, I create classes as blueprints and make multiple objects from them. 

This is super useful for real projects. For example, managing a school with hundreds of students, teachers, or even an e-commerce site with products and orders. OOP makes the code reusable, easier to maintain, and scalable.

### Defining a Class and Constructor
A `class` is like a template or blueprint. When I create an object (instance), the `__init__` method (constructor) runs automatically to initialize its properties. `self` represents the current object, so each instance has its own data. I don\'t pass `self` manually - Python handles it.

```python
class Student:
    def __init__(self, name, roll_no, marks, grade="Pending"):
        self.name = name          # instance variable
        self.roll_no = roll_no
        self.marks = marks
        self.grade = grade

# Creating objects
s1 = Student("Priya Sharma", 101, 92)
s2 = Student("Rahul Verma", 102, 78, "B")

print(s1.name, s1.marks, s1.grade)
print(s2.name, s2.marks, s2.grade)
```

### Methods in Classes
Methods are functions defined inside the class. They can access and modify the object\'s data using `self`. This keeps everything neatly together.

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    
    def calculate_grade(self):
        if self.marks >= 90:
            return "A+"
        elif self.marks >= 75:
            return "A"
        elif self.marks >= 60:
            return "B"
        else:
            return "C"
    
    def is_passing(self):
        return self.marks >= 40

s1 = Student("Priya", 92)
s2 = Student("Rahul", 55)

print(f"{s1.name} got grade: {s1.calculate_grade()}")
print(f"{s2.name} is passing: {s2.is_passing()}")
```

### The __str__ Method
By default, printing an object gives a memory address. Defining `__str__` lets me customize how the object is displayed when printed or converted to string.

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks
    
    def __str__(self):
        return f"Student: {self.name} (Roll: {self.roll_no}, Marks: {self.marks})"

s = Student("Ananya", 105, 87)
print(s)  # Uses __str__ automatically
```

### Encapsulation - Protecting Data
Encapsulation hides internal data and exposes it only through methods. In Python, we use single underscore `_` for protected and double `__` for private (name mangling). This prevents accidental changes and adds validation.

```python
class BankAccount:
    def __init__(self, owner, initial_balance):
        self.owner = owner
        self.__balance = initial_balance  # private
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance: {self.__balance}")
        else:
            print("Invalid deposit amount")
    
    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}. Remaining: {self.__balance}")
        else:
            print("Insufficient balance or invalid amount")
    
    def get_balance(self):
        return self.__balance

acc = BankAccount("Me", 5000)
acc.deposit(1500)
acc.withdraw(7000)  # Blocked
acc.withdraw(2000)
print("Final balance:", acc.get_balance())
```

### Inheritance - Reusing Code
Inheritance allows a child class to inherit attributes and methods from a parent class. Use `super()` to call the parent\'s constructor. This promotes code reuse.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def introduce(self):
        return f"Hello, I am {self.name}, {self.age} years old."

class Student(Person):
    def __init__(self, name, age, roll_no, marks):
        super().__init__(name, age)  # Call parent constructor
        self.roll_no = roll_no
        self.marks = marks
    
    def introduce(self):  # Overriding
        base = super().introduce()
        return f"{base} I am a student with roll no {self.roll_no}."

student = Student("Arjun", 17, 203, 85)
print(student.introduce())
print("Marks:", student.marks)
```

### Polymorphism and Method Overriding
Polymorphism lets objects of different classes respond to the same method call differently. Overriding is redefining a method in the child class.

```python
class Shape:
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius ** 2

# Polymorphism in action
shapes = [Rectangle(5, 3), Circle(4)]
for shape in shapes:
    print(f"Area: {shape.area()}")
```

### File Input/Output (I/O)
Python makes it easy to work with files using `open()`. Modes: `\'r\'` (read), `\'w\'` (write/overwrite), `\'a\'` (append). Always use `with` statement for automatic cleanup.

```python
# Writing to a file
students_data = [
    "Priya Sharma,101,92\n",
    "Rahul Verma,102,78\n",
    "Ananya Singh,103,95\n"
]

with open("class_students.txt", "w") as file:
    file.writelines(students_data)

print("Student data written to file.")
```

```python
# Reading the file
with open("class_students.txt", "r") as file:
    content = file.read()
    print("File contents:")
    print(content)
```

```python
# Appending new data
with open("class_students.txt", "a") as file:
    file.write("Arjun Kumar,104,88\n")

# Reading again with lines
print("\nUpdated student list:")
with open("class_students.txt", "r") as file:
    for line in file:
        print(line.strip())
```

### Recap & Best Practices
- **Class & Object**: Blueprint vs actual instance
- **__init__ & self**: Setup data for each object
- **Methods**: Behavior tied to the object
- **__str__**: Nice string representation
- **Encapsulation**: Hide data, control access
- **Inheritance**: Extend existing classes with `super()`
- **Polymorphism**: Same interface, different implementations
- **File I/O**: Use `with open(...)` for safety

OOP + File handling is the foundation for building real applications like management systems, games, or data processors.
