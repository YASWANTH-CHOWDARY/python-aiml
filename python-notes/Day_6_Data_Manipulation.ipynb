## LOADING DATA

We use pd.read_csv() to load a CSV file into a DataFrame.

A DataFrame is basically a table with rows and columns, just like an Excel sheet.

## PANDAS SERIES

A Series is a 1-dimensional structure in Pandas.
Think of it as a single column with an index attached to each value.

When you select one column from a DataFrame, you get a Series back.

```python
# creating a basic Series with default index (0, 1, 2, 3)
marks = pd.Series([80, 90, 70, 95])
marks
```

You can also create a Series with a custom index instead of the default 0, 1, 2.
This makes it easy to label your data meaningfully.

```python
# Series with custom index
marks = pd.Series(
    [80, 90, 70],
    index=['Lahari', 'Bharathi', 'Pranav']
)
marks
```

## SERIES ARITHMETIC

When you do arithmetic on two Series, Pandas automatically matches values by their index labels.

If a label exists in one Series but not the other, the result for that label will be NaN (Not a Number).
This is called index alignment.

```python
s1 = pd.Series([10, 20, 30], index=['A', 'B', 'C'])
s2 = pd.Series([5, 15, 25],  index=['A', 'B', 'D'])

# A matches A, B matches B
# C is only in s1 so result is NaN
# D is only in s2 so result is NaN
s1 + s2
```

## CREATING A DATAFRAME FROM SCRATCH

You can create a DataFrame from a dictionary.
Each key becomes a column name and the list of values becomes the data for that column.

All lists must have the same length.

```python
data = {
    'Student': ['Lahari', 'Pranav', 'Bharathi', 'Bannuru', 'Srikar', 'Teja', 'Vinay'],
    'Math':    [85, 92, 78, 95, 88, 81, 94],
    'Science': [90, 89, 85, 98, 80, 87, 92],
    'English': [95, 98, 90, 99, 92, 94, 89]
}

df = pd.DataFrame(data)
print(df)
```

## LOC AND ILOC

loc and iloc are two ways to select rows from a DataFrame.

loc  - selects rows based on the INDEX LABEL
iloc - selects rows based on the INTEGER POSITION (like a list index starting from 0)

When the index is the default (0, 1, 2...) both give the same result.
But when the index is custom (like names), they behave differently.

```python
# loc - returns row with label 1
df.loc[1]
```

```python
# iloc - returns the second row (position 1, 0-indexed)
df.iloc[1]
```

## ADDING A CALCULATED COLUMN

You can create a new column by doing calculations using existing columns.
The operation is done row by row automatically.

```python
# calculate the average of Math and Science and store it in a new column
df['Average'] = (df['Math'] + df['Science']) / 2
df
```

## BOOLEAN FILTERING

Boolean filtering lets you filter rows based on a condition.

When you write df['Math'] > 85, it returns a True/False value for each row.
When you pass that inside df[], it returns only the rows where the condition is True.

This is one of the most useful operations in Pandas.

```python
# students scoring more than 85 in Math
df[df['Math'] > 85]
```

You can combine multiple conditions using & (AND) and | (OR).
Each condition must be wrapped in its own parentheses.

```python
# multiple conditions - Math > 80 AND Science > 85
df[
    (df['Math'] > 80)
    &
    (df['Science'] > 85)
]
```

After filtering rows, you can also select only specific columns from the result.

```python
# selecting specific columns after filtering
df[df['Math'] > 80][['Student', 'Math']]
```

## HANDLING MISSING VALUES

Real world data almost always has missing values. Pandas represents missing values as NaN (Not a Number).

We need to handle them before doing any analysis because NaN values can mess up calculations.

Two main ways to handle missing values:
1. dropna() - remove the rows or columns that have missing values
2. fillna() - fill the missing values with something (a number, mean, median etc)

```python
# creating a dataframe with some missing values using np.nan
data = {
    'Student': ['Lahari', 'Pranav', 'Bharathi', 'Bannuru', 'Srikar', 'Teja', 'Vinay'],
    'Math':    [85, 92, 78, 95, 88, 81, np.nan],
    'Science': [90, 89, 85, np.nan, 80, 87, 92],
    'English': [95, np.nan, 90, 99, 92, np.nan, 89]
}

df = pd.DataFrame(data)
print(df)
```

## DROPNA

dropna() by default removes any row that has at least one missing value.
Only rows having no missing values at all will remain.

```python
# removes rows with any missing values
df.dropna()
```

axis=1 means you are applying the operation on columns instead of rows.
So dropna(axis=1) drops any column that has even one missing value.

```python
# axis=1 means columns - drops columns that have any null value
df.dropna(axis=1)
```

thresh parameter lets you control how many non-null values a row must have to be kept.
For example thresh=3 means keep only rows that have at least 3 non-null values.

```python
# thresh=3 means keep rows that have minimum 3 non-null values
df.dropna(thresh=3)
```

## FILLNA

fillna() fills missing values instead of removing them.
You can fill with a constant value like 0 or a string like 'Not Available'.

A smarter approach is to fill with the mean or median of that column so you dont distort the data too much.

```python
# fill all missing values with a constant string
df.fillna('Not Available')
```

```python
# fill missing marks in English with the average English marks
df['English'].fillna(
    df['English'].mean()
)
```

## GROUPBY

groupby() is used to group rows together based on the values in a column, and then apply an aggregation on each group.

Think of it like this - if you have student data with departments, groupby('Department') will split the data into separate groups for each department. Then you can calculate the average, sum, max etc. for each group.

The logic is:
1. Split the data into groups
2. Apply a function (mean, sum, count etc)
3. Combine the results

```python
students = pd.DataFrame({
    'Student':     ['Lahari', 'Pranav', 'Bharathi', 'Bannuru', 'Srikar', 'Teja', 'Vinay'],
    'Department':  ['Math', 'Biology', 'Math', 'English', 'Math', 'Biology', 'English'],
    'Total_Marks': [750, 650, 820, 680, 700, 780, 840]
})
```

```python
# groupby alone just creates a GroupBy object, nothing is shown yet
students.groupby('Department')
```

```python
# average marks by department
students.groupby('Department').mean(numeric_only=True)
```

```python
# total marks by department
students.groupby('Department').sum(numeric_only=True)
```

```python
# maximum marks by department
students.groupby('Department').max(numeric_only=True)
```

```python
# count of students in each department
students.groupby('Department').count()
```

## CONCAT

concat is used to combine two or more DataFrames together.

By default it stacks them vertically (one below the other). This is useful when you have data split across multiple files or batches.

With axis=1 it stacks them horizontally (side by side as new columns). This is useful when you want to add new columns to a DataFrame.

```python
team1 = pd.DataFrame({
    'student':    ['Lahari', 'Pranav', 'Teja'],
    'Total_Marks':[750, 650, 780]
})
```

```python
team2 = pd.DataFrame({
    'student':    ['Bharathi', 'Srikar', 'Vinay', 'Bannuru'],
    'Total_Marks':[820, 700, 840, 680]
})
```

```python
# stacking vertically - one below the other
# notice the index still shows original values (0,1,2 repeated)
pd.concat([team1, team2])
```

```python
# stacking horizontally - side by side as new columns
dept = pd.DataFrame({
    'Department': ['Math', 'Biology', 'English']
})

pd.concat([team1, dept], axis=1)
```

## MERGE - JOINING DATAFRAMES

merge is used to join two DataFrames based on a common column. It works exactly like SQL joins.

The 'on' parameter tells which column to use for matching.
The 'how' parameter tells what type of join to do.

Types of joins:

inner  - keeps only rows where the key exists in BOTH DataFrames
left   - keeps all rows from the left DataFrame, and only matching rows from the right. Non-matching rows from right get NaN
right  - keeps all rows from the right DataFrame, and only matching rows from the left. Non-matching rows from left get NaN
outer  - keeps all rows from BOTH DataFrames. NaN where there is no match on either side

```python
students1 = pd.DataFrame({
    'studID':  [1, 2, 3],
    'student': ['Lahari', 'Pranav', 'Teja']
})
```

```python
students2 = pd.DataFrame({
    'studID':      [1, 2, 4],
    'Total_Marks': [750, 650, 780]
})
```

```python
# inner join - only studID 1 and 2 are common in both, so only they appear
pd.merge(
    students1,
    students2,
    how='inner',
    on='studID'
)
```

```python
# left join - all rows from students1 are kept
# studID 3 (Teja) exists in students1 but not in students2, so Total_Marks is NaN
pd.merge(
    students1,
    students2,
    how='left',
    on='studID'
)
```

```python
# right join - all rows from students2 are kept
# studID 4 exists in students2 but not in students1, so student name is NaN
pd.merge(
    students1,
    students2,
    how='right',
    on='studID'
)
```

```python
# outer join - all rows from both DataFrames are kept
# NaN appears wherever there is no match
pd.merge(
    students1,
    students2,
    how='outer',
    on='studID'
)
```
