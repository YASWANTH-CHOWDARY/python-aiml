## Loops

A loop allows us to execute a block of code multiple times without rewriting the same statements. Loops reduce repetition and make programs easier to maintain.

## For Loop

A for loop is used when we want to process every item in a sequence or run code a known number of times.

```python
courses = ['Python','SQL','Machine Learning']
for course in courses:
    print(course)
```

## range()

The range() function generates a sequence of numbers. The stop value is not included.

```python
for week in range(1,5):
    print('Week', week)
```

## Running Total Example

A variable can be updated during each iteration to keep track of a cumulative result.

```python
total = 0
for score in range(1,6):
    total += score
print('Total =', total)
```

## While Loop

A while loop runs as long as its condition remains True. It is useful when the number of iterations is not known beforehand.

```python
energy = 100
while energy > 0:
    print('Energy:', energy)
    energy -= 25
```

## User Controlled Loop

Many applications keep asking for input until the user enters a valid response.

```python
answer = ''
while answer != 'quit':
    answer = input('Type quit to exit: ')
print('Program Ended')
```

## Counting Backwards

A negative step allows range() to move backwards.

```python
for num in range(5,0,-1):
    print(num)
```

## break Statement

break immediately terminates the loop and moves execution to the first statement after the loop.

```python
for room in range(1,11):
    if room == 7:
        print('Target Room Found')
        break
    print('Checking Room', room)
```

## continue Statement

continue skips the current iteration and moves directly to the next iteration.

```python
for number in range(1,11):
    if number % 3 == 0:
        continue
    print(number)
```

## Strings

A string is a sequence of characters enclosed in quotes. Strings can contain letters, numbers, spaces and symbols.

```python
university = 'University at Buffalo'
print(university)
```

## len()

len() returns the total number of characters in a string.

```python
university = 'University at Buffalo'
print(len(university))
```

## String Indexing

Indexing accesses a single character using its position. Positive indexing starts from the left and negative indexing starts from the right.

```python
city = 'Buffalo'
print(city[0])
print(city[-1])
```

## String Slicing

Slicing extracts a portion of a string using start and end positions.

```python
course = 'MachineLearning'
print(course[0:7])
print(course[::-1])
```

## String Multiplication

The * operator can repeat a string multiple times.

```python
print('AI ' * 3)
```

## upper() and lower()

These methods convert text to uppercase or lowercase.

```python
name = 'Yaswanth'
print(name.upper())
print(name.lower())
```

## split()

split() breaks a string into smaller pieces and returns a list.

```python
skills = 'Python,SQL,AI,ML'
print(skills.split(','))
```

## count()

count() returns how many times a substring appears in a string.

```python
text = 'mississippi'
print(text.count('s'))
```

## find()

find() returns the position of the first occurrence. If not found, it returns -1.

```python
email = 'yaswanth@buffalo.edu'
print(email.find('@'))
print(email.find('#'))
```

## expandtabs()

expandtabs() converts tab characters into spaces.

```python
text = 'Name\tCourse\tGrade'
print(text.expandtabs())
```

## isalnum()

Returns True if all characters are letters or numbers.

```python
print('Python123'.isalnum())
```

## isalpha()

Returns True if all characters are alphabetic.

```python
print('Python'.isalpha())
```

## islower() and isupper()

Used to check whether letters are lowercase or uppercase.

```python
print('python'.islower())
print('PYTHON'.isupper())
```

## isspace()

Returns True if the string contains only whitespace characters.

```python
print('   '.isspace())
```

## istitle()

Checks whether each word starts with an uppercase letter followed by lowercase letters.

```python
print('Machine Learning'.istitle())
```

## endswith()

Returns True if a string ends with a specified suffix.

```python
filename = 'notes.ipynb'
print(filename.endswith('.ipynb'))
```

## partition()

partition() splits a string into exactly three parts around the first occurrence of a separator.

```python
text = 'Python-AI'
print(text.partition('-'))
```

## Key Takeaways

- Use for loops when iterations are known.
- Use while loops when execution depends on a condition.
- break exits a loop.
- continue skips the current iteration.
- Strings are immutable.
- Indexing accesses one character while slicing accesses multiple characters.
