This notebook demonstrates multiple ways to reverse strings in Python.

Topics Covered:

- For Loop
- For Loop with range()
- While Loop
- reversed() Function
- Reverse Each Word
- Reverse Word Order
- Reverse Words and Characters
- Partial Reversal

## Method 1: Using Slicing

The simplest and most Pythonic approach. The step value `-1` tells Python to traverse the string from right to left.

```python
word = 'YASWANTH'
print(word[::-1])
```

## Method 2: Using a For Loop

Each new character is added to the beginning of the result string, gradually building the reversed version.

```python
word = 'YASWANTH'
result = ''

for ch in word:
    result = ch + result

print(result)
```

## Method 3: Using range()

The loop starts from the last index and moves backward until index 0.

```python
word = 'YASWANTH'

for i in range(len(word)-1, -1, -1):
    print(word[i], end='')
```

## Method 4: Using a While Loop

A while loop gives more control over the index movement while traversing the string backwards.

```python
word = 'YASWANTH'
index = len(word)-1

while index >= 0:
    print(word[index], end='')
    index -= 1
```

## Method 5: Using reversed()

The reversed() function returns characters in reverse order. join() combines them into a single string.

```python
word = 'YASWANTH'
result = ''.join(reversed(word))
print(result)
```

## Method 6: Store Reversed String in a Variable

Build a new string one character at a time using reverse indexing.

```python
word = 'YASWANTH'
result = ''

for i in range(len(word)-1, -1, -1):
    result += word[i]

print(result)
```

## Method 7: Reverse Each Word in a Sentence

Each word is reversed individually while the word order remains unchanged.

```python
sentence = 'Python SQL AI'

for word in sentence.split():
    print(word[::-1], end=' ')
```

## Method 8: Reverse the Order of Words

Characters stay the same, but the positions of the words are reversed.

```python
sentence = 'Python SQL AI'
words = sentence.split()

print(' '.join(words[::-1]))
```

## Method 9: Reverse Both Words and Characters

First reverse the word positions and then reverse each individual word.

```python
sentence = 'Python SQL AI'
words = sentence.split()[::-1]

for word in words:
    print(word[::-1], end=' ')
```

## Method 10: Reverse Only the First Few Characters

This technique is useful when only a specific portion of the string needs to be reversed.

```python
word = 'BUFFALO'
result = word[:4][::-1] + word[4:]
print(result)
```

## Method 11: Reverse Only Alphabets

Numbers and symbols are ignored. Only alphabetic characters participate in the reversal.

```python
text = 'Yaswanth2026Python'
letters = ''

for ch in text:
    if ch.isalpha():
        letters = ch + letters

print(letters)
```

## Method 12: Using Recursion

The function keeps calling itself on smaller portions of the string until the base condition is reached.

```python
def reverse_string(text):
    if len(text) == 0:
        return text

    return reverse_string(text[1:]) + text[0]

print(reverse_string('YASWANTH'))
```

## Method 13: Using a Stack

A stack follows Last-In-First-Out (LIFO). Characters are popped in reverse order.

```python
word = 'YASWANTH'
stack = list(word)
result = ''

while stack:
    result += stack.pop()

print(result)
```

## Key Takeaways

- Slicing is the shortest and most commonly used approach.
- reversed() is a built-in Python solution.
- Loops help understand the reversal logic step by step.
- Recursion demonstrates function call behavior.
- Stack-based reversal follows the LIFO principle used in data structures.
