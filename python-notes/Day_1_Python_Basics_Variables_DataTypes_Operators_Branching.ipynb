
# Day 1 - Python Basics, Variables, Data Types, Operators and Branching

These notes cover the fundamental concepts of Python programming. The goal is to understand both the syntax and what happens behind the scenes when Python executes code.


## What is Python?

Python is a high-level, interpreted programming language that focuses on readability and simplicity.

Python is widely used in Web Development, Data Science, AI/ML, Automation, Cloud Computing and Cyber Security.

```python
print("Hello World")
```

## Print Function

The `print()` function is used to display output on the screen. It is one of the most commonly used functions while learning Python.

```python
print("Hello Yaswanth")
print(17 + 9)
```

## Variables

A variable is a container used to store data. Internally, variables act as references to objects stored in memory rather than storing the actual value themselves.

```python
name = "Yaswanth"
age = 24
print(name)
print(age)
```

## Variable Reassignment

A variable can be assigned a new value at any time.

When a variable is reassigned, Python updates the reference to point to the new object. When you print the variable, Python returns the most recently assigned value.

```python
city = "Buffalo"
print(city)

city = "Williamsville"
print(city)
```

## Integer (int)

Integers are whole numbers without decimal points. They can be positive, negative or zero.

```python
x = 10
print(type(x))
```

## Float (float)

Floats are numbers that contain decimal values. Python automatically uses float when a number has a decimal point.

```python
x = 10.5
print(type(x))
```

## Boolean (bool)

Boolean values represent only two states: True and False. They are commonly used in decision making and conditions.

```python
is_logged_in = True
print(is_logged_in)
print(type(is_logged_in))
```

## Strings (str)

A string is a sequence of characters enclosed in single or double quotes.

```python
name = "Python"
print(type(name))
```

## Type Conversion

Type conversion is the process of changing data from one type to another.

Python supports implicit conversion automatically and explicit conversion using functions like int(), float() and str().

```python
x = 5
y = 2.5
print(x + y)

age = "24"
age = int(age)
print(type(age))
```

## Operators

Operators are special symbols used to perform operations on values and variables.

```python
a = 10
b = 3

print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a%b)
print(a**b)
```

## Comparison Operators

Comparison operators compare two values and always return either True or False.

```python
print(10 > 5)
print(10 == 5)
print(10 != 5)
```

## Logical Operators

Logical operators are used to combine multiple conditions together.

```python
print(True and False)
print(True or False)
print(not True)
```

## Input Function

The input() function allows users to provide input while the program is running.

Important: input() always returns a string by default, even if the user enters a number.

```python
name = input("Enter Name: ")
print(name)
```

## Converting User Input

Since input() returns a string, we often convert it to another data type before performing calculations.

```python
age = int(input("Enter Age: "))
print(type(age))
```

## f-Strings

f-strings provide a clean and readable way to insert variables inside text.

```python
name = "Yaswanth"
salary = 90000
print(f"Hello {name}, your salary is {salary}")
```

## format() Method

Before f-strings became popular, the format() method was commonly used for string formatting.

```python
name = "Yaswanth"
print("Hello {}".format(name))
```

## Branching (if, elif, else)

Branching allows a program to make decisions based on conditions.

Python evaluates conditions from top to bottom and stops at the first condition that is True.

```python
marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
else:
    print("Grade C")
```

## Nested If

A nested if statement is an if statement placed inside another if statement. It is useful when multiple conditions depend on one another.

```python
age = 22
has_license = True

if age >= 18:
    if has_license:
        print("Can Drive")
    else:
        print("License Required")
```

## Key Takeaways

- Python is dynamically typed.
- Variables can be reassigned.
- Input returns string by default.
- Comparison operators return Boolean values.
- Python uses indentation to define code blocks.
- Python stops at the first True condition in an if-elif-else chain.
