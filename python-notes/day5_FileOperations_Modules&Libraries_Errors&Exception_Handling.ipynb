# PART 1 — FILE OPERATIONS

Working with files in Python basically means: create files, write stuff into them, read them back, add more stuff, or copy them.

The main tool for this is the built-in `open()` function:

```python
file = open("file_path", "mode")
```

### The different modes you need to know:
- `"w"` → **write** mode — creates the file or wipes existing content (careful with this!)
- `"r"` → **read** mode — reads an existing file
- `"a"` → **append** mode — adds content at the end WITHOUT deleting old stuff
- `"r+"` → **read + write** — both at once, but file must already exist

### Better way — always use `with open()`
It automatically closes the file when done. No need to manually call `file.close()`.

```python
with open("file_path", "mode") as file:
    # do stuff here
```

## Writing to a File

Use `"w"` mode. Creates the file if it doesn't exist.  
**Warning:** if the file already has content, it gets wiped. Overwrite city.

````python
file_path = "sample.txt"

with open(file_path, "w") as file:
    file.write("Hello!\n")
    file.write("Welcome to Python file operations.\n")
    file.write("Today we are learning how to write into a text file.\n")

print("File written successfully.")
````

## Reading from a File

Use `"r"` mode to read an existing file.  
**Heads up:** if the file doesn't exist → Python throws a `FileNotFoundError`.

````python
file_path = "sample.txt"

with open(file_path, "r") as file:
    content = file.read()

print(content)
````

## Reading Line by Line (for loop)

Don't want to read the whole file at once? Use a for loop.  
Useful when files are huge. Each iteration gives you one line.

**Small annoyance:** you'll get extra blank lines because each line already has `\n`  
and `print()` adds another. Fix it with `.strip()`.

````python
file_path = "sample.txt"

with open(file_path, "r") as file:
    for line in file:
        print(line.strip())  # .strip() removes the extra newline
````

## Appending to a File

Use `"a"` mode to add new content at the END of the file.  
Old content stays safe — nothing gets deleted. 

````python
file_path = "sample.txt"

with open(file_path, "a") as file:
    file.write("This line is added using append mode.\n")
    file.write("Append mode adds content at the end.\n")

print("New content appended successfully.")
````

## Reading Only the First N Characters

Don't need the whole file? Just read the first N characters with `file.read(N)`.  
Note: spaces count as characters too.

````python
file_path = "sample.txt"

with open(file_path, "r") as file:
    first_12 = file.read(12)

print("First 12 chars:", first_12)
````

## Reading Line by Line with `.readline()`

`.readline()` reads ONE line at a time.  
Each call moves to the next line automatically.  
Again, use `.strip()` to clean up the extra newline characters.

````python
file_path = "sample.txt"

with open(file_path, "r") as file:
    line1 = file.readline().strip()
    line2 = file.readline().strip()
    line3 = file.readline().strip()

print(line1)
print(line2)
print(line3)
````

## Read ALL Lines into a List — `.readlines()`

Reads every line and stores each one as an item in a list.  
Super handy when you want to process lines individually later.

````python
file_path = "sample.txt"

with open(file_path, "r") as file:
    lines_list = file.readlines()

print(lines_list)  # each line is a list item, \n included
````

## Copying a File

No built-in copy command here — we do it manually:  
1. Read the source file  
2. Write that content into a new destination file  

Simple but effective.

````python
source_path = "sample.txt"
destination_path = "copy_sample.txt"

with open(source_path, "r") as source_file:
    content = source_file.read()

with open(destination_path, "w") as destination_file:
    destination_file.write(content)

print("File copied successfully.")
````

## `r+` Mode — Read AND Write Together

Things to keep in mind about `r+`:
- File **must already exist** (won't create a new one)
- Writing starts from the **beginning** of the file
- Can overwrite existing characters
- Use `file.seek(0)` to move cursor back to the start

````python
file_path = "sample.txt"

with open(file_path, "r+") as file:
    content = file.read()
    print("Before writing:")
    print(content)

    file.seek(0)  # go back to the beginning
    file.write("Updated first line!\n")

print("r+ operation done.")
````

# PART 2 — MODULES & LIBRARIES

As your programs get bigger, you don't want to write everything from scratch.  
Python has **modules** and **libraries** for that.

**Quick definitions:**
- **Module** → a single `.py` file with reusable code
- **Package** → a folder of multiple modules (has an `__init__.py`)
- **Library** → a big collection of packages and modules

Import with: `import module_name`

## `math` Module

Built-in math stuff — square roots, rounding, factorials, pi, etc.

````python
import math

# Square root
print(math.sqrt(64))        # 8.0

# ceil() rounds UP, floor() rounds DOWN
print(math.ceil(4.2))       # 5
print(math.floor(4.8))      # 4

# Factorial: 5! = 5 x 4 x 3 x 2 x 1
print(math.factorial(5))    # 120

# Pi for area of circle
radius = 5
area = math.pi * radius ** 2
print(area)                 # 78.53...
````

## `datetime` Module

All things date and time — current date/time, custom dates, calculating differences between dates.

````python
import datetime

# Current date and time
current = datetime.datetime.now()
print(current)

# Just the date
today = datetime.date.today()
print(today)

# Just the time
current_time = datetime.datetime.now().time()
print(current_time)

# Custom date
birthday = datetime.date(2000, 8, 9)
print(birthday)

# Days between two dates
difference = today - birthday
print(difference.days, "days old")
````

## `sys` Module

Interacts with the Python interpreter itself.  
Useful for checking Python version, platform, etc.

````python
import sys

print(sys.version)    # Python version
print(sys.platform)   # OS platform (linux, win32, darwin, etc.)
````

## `statistics` Module

Math stats stuff — mean, median, mode, standard deviation.  
Super useful for data analysis and ML projects.

````python
import statistics

marks = [85, 90, 78, 92, 88]
print("Mean:", statistics.mean(marks))       # average

numbers = [10, 20, 30, 40, 50]
print("Median:", statistics.median(numbers)) # middle value

numbers2 = [10, 20, 20, 30, 40]
print("Mode:", statistics.mode(numbers2))    # most frequent

marks2 = [80, 85, 90, 95, 100]
print("Std Dev:", statistics.stdev(marks2))  # standard deviation
````

# PART 3 — ERRORS & EXCEPTION HANDLING

An error = something that makes your program crash and stop.  

**Without handling:**
- Program just crashes
- User sees an ugly traceback
- Nothing after the error runs

**With handling:**
- Program keeps running
- User sees a friendly message
- Your app is way more reliable

### Common error types:
- `ZeroDivisionError` → dividing by 0
- `ValueError` → wrong type of value (like `int("abc")`)
- `IndexError` → index out of range in a list
- `KeyError` → key doesn't exist in a dictionary
- `FileNotFoundError` → file doesn't exist

## `try` / `except` — The Basic Structure

`try` block = code that might break  
`except` block = what to do when it breaks  

**Best practice:** always name the specific exception type instead of a bare `except:`

````python
# ZeroDivisionError
try:
    print(10 / 0)
except ZeroDivisionError:
    print("Can't divide by zero!")

print("Program keeps going...")
````

````python
# ValueError from bad user input
try:
    age = int(input("Enter your age: "))
    print("Age:", age)
except ValueError:
    print("Please enter a number only.")
````

````python
# File not found
try:
    with open("student.txt", "r") as file:
        print(file.read())
except FileNotFoundError:
    print("That file doesn't exist.")
````

## Raising Your Own Exceptions

You can manually throw an error with `raise Exception("your message")`.  
Use it when you want to enforce your own rules (like checking balances).

````python
try:
    balance = 5000
    amount = int(input("Enter withdrawal amount: "))

    if amount > balance:
        raise Exception("Insufficient Balance")

    print("Withdrawal Successful")

except Exception as e:
    print("Transaction Failed:", e)
````

## `finally` Block

`finally` **always runs** — whether there was an error or not.  

Common uses:
- Closing files or database connections
- Releasing resources
- Logging

````python
try:
    print(10 / 0)
except ZeroDivisionError:
    print("Error occurred.")
finally:
    print("Finally block always runs — error or not.")
````

````python
# Real-ish example
try:
    marks = int(input("Enter marks: "))
    percentage = marks / 500 * 100
    print("Percentage:", percentage)

except ValueError:
    print("Please enter valid marks.")

finally:
    print("Thanks for using the Result Calculator.")
````

## Exception Flow Diagram

```
           START
             |
           [try]
             |
       Error occurs?
       /           \
     YES            NO
      |              |
  [except]      (skip except)
      |              |
      +---->[finally]<----+
                 |
               END
```

Simple rule:
- `try` → attempt the risky code
- `except` → handle it if something goes wrong
- `finally` → runs no matter what (cleanup code goes here)
