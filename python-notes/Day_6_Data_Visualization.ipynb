## What is Data Visualization?

Data Visualization is the graphical representation of data using charts, graphs, and plots.

When we have a large dataset with hundreds or thousands of rows, it becomes very difficult to understand what the data is saying just by reading numbers. Visualization solves this problem by converting numbers into a visual form that our brain can understand quickly.

For example, if you have the marks of 100 students, you cannot easily tell who scored highest or what the average is just by reading the list. But if you plot it as a bar chart, the highest and lowest values are immediately visible.

Why we use Data Visualization:

- To identify trends in data (is sales going up or down over time?)
- To compare values across categories (which student scored highest?)
- To find patterns (do tips increase on weekends?)
- To detect outliers (is one bill unusually high compared to others?)
- To communicate findings clearly to others
- To make faster and better decisions

A single well-made graph can replace an entire page of numbers and tables.

## Libraries Used

We need four libraries for data visualization in Python.

NumPy - Used for numerical operations, mathematical functions, and working with arrays. It is the foundation that most other data science libraries are built on.

Pandas - Used for data manipulation. It lets us create and work with DataFrames, which are table-like structures. Most real-world datasets are loaded and handled using Pandas.

Seaborn - A visualization library built on top of Matplotlib. It produces cleaner and more attractive statistical plots with less code. It also comes with built-in sample datasets.

Matplotlib - The core plotting library in Python. It gives us functions like plt.title(), plt.xlabel(), plt.show() which we use to control the appearance of our graphs. Seaborn and Pandas both use Matplotlib under the hood.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Creating a DataFrame

A DataFrame is a two-dimensional data structure with rows and columns, similar to an Excel spreadsheet or a SQL table.

We create a DataFrame by passing a Python dictionary to pd.DataFrame(). In the dictionary:
- Each key becomes a column name
- Each list becomes the data in that column
- All lists must have the same length

Here we are creating a student marks dataset with 7 students and their scores in Math, Science, and English.

```python
data = {
    "Student": ["Lahari", "Pranav", "Bharathi", "Bannuru", "Srikar", "Teja", "Vinay"],
    "Math":    [85, 92, 78, 95, 88, 81, 94],
    "Science": [90, 89, 85, 98, 80, 87, 92],
    "English": [95, 98, 90, 99, 92, 94, 89]
}

df = pd.DataFrame(data)
print(df)
```

## Data Types - DataFrame vs Series

When working with Pandas, there are two important data structures to know.

DataFrame - A two-dimensional table. It has both rows and columns. When you print df, you get a DataFrame.

Series - A one-dimensional array. When you select a single column from a DataFrame using df["column_name"], the result is a Series. It is basically a list with an index attached to each value.

Understanding the difference matters because some functions work only on a Series, and some work only on a DataFrame.

```python
# Check type of the full DataFrame
print(type(df))
```

```python
# Check type of a single column
print(type(df["Math"]))
```

```python
# Print a single column (Series)
print(df["Math"])
```

## The plot() Function

The plot() function is built into Pandas and is the simplest way to create a graph from your data. When called on a Series or DataFrame, it automatically creates a line chart by default.

Syntax:
    data.plot()

You can pass additional parameters to control the type of chart, the axes, colors, and more.

plt.show() is needed at the end to actually display the graph. Without it, the graph may not render in some environments.

```python
# Simplest plot - just call plot() on a column
df["Math"].plot()
plt.show()
```

```python
# Plotting all numeric columns at once
# Pandas automatically skips non-numeric columns like "Student"
df.plot()
plt.show()
```

## Adding Labels to Graphs

A graph without labels is hard to understand. We always add at minimum a title and axis labels so anyone reading the graph knows what they are looking at.

plt.title() - Adds a title at the top of the graph
plt.xlabel() - Adds a label below the x-axis (horizontal axis)
plt.ylabel() - Adds a label to the left of the y-axis (vertical axis)

These functions come from Matplotlib and are called after the plot() call but before plt.show().

```python
# Plot Math scores with proper labels
df.plot(x="Student", y="Math")

plt.title("Math Scores of Students")
plt.xlabel("Student")
plt.ylabel("Math Marks")

plt.show()
```

```python
# Plot Science scores with labels
df.plot(x="Student", y="Science")

plt.title("Science Scores of Students")
plt.xlabel("Student")
plt.ylabel("Science Marks")

plt.show()
```

```python
# Plot all three subjects together with labels
df.plot(x="Student", y=["Math", "Science", "English"])

plt.title("All Subject Scores")
plt.xlabel("Student")
plt.ylabel("Marks")

plt.show()
```

## Plotting Total Bills

## iloc() - Integer Location Indexing

iloc stands for integer location. It is used to select specific rows and columns from a DataFrame using position numbers, not column names.

Syntax:
    df.iloc[row_selection, column_selection]

Row and column selection can be:
- A single number: iloc[0] means row 0
- A slice: iloc[0:5] means rows 0 to 4 (5 is excluded)
- A colon alone: iloc[:] means all rows or all columns
- A list: iloc[:, [0, 2]] means all rows, columns 0 and 2 only

Note: In Python slicing, the end index is always excluded. So 0:5 gives positions 0, 1, 2, 3, 4.

```python
# Select first 5 rows, all columns
print(tips.iloc[0:5])
```

```python
# Select all rows, first 2 columns only (total_bill and tip)
print(tips.iloc[:, 0:2].head())
```

```python
# Select first 10 rows of total_bill column only
print(tips.iloc[0:10, [0]])
```

```python
# Select a single row by position
print(tips.iloc[0])
```

```python
# Select rows 5 to 10, columns 0 and 1
print(tips.iloc[5:11, 0:2])
```

## Plotting Multiple Columns Together

When we use iloc to select multiple columns and then call plot(), all selected columns are drawn on the same graph with different colors. This makes it easy to compare them visually.

Here we plot total_bill and tip on the same graph so we can see how tip relates to the bill amount across all customers.

```python
# Plot total_bill and tip together
tips.iloc[:, 0:2].plot()

plt.title("Total Bill and Tip")
plt.xlabel("Customer Index")
plt.ylabel("Amount ($)")

plt.show()
```

```python
# Plot first 10 customers only for a cleaner view
tips.iloc[0:10, 0:2].plot()

plt.title("Total Bill and Tip - First 10 Customers")
plt.xlabel("Customer Index")
plt.ylabel("Amount ($)")

plt.show()
```

## Bar Plot

A bar chart shows each value as a vertical bar. The height of the bar represents the value.

Bar charts are better than line charts when:
- You are comparing separate, distinct categories
- The data is not continuous (no meaningful connection between points)
- You want to highlight the difference in magnitude between values

To create a bar chart, we pass kind="bar" inside the plot() function.

```python
# Bar chart of total bill for first 10 customers
tips.iloc[0:10, [0]].plot(
    kind="bar",
    legend=False
)

plt.title("Total Bill - First 10 Customers")
plt.xlabel("Customer Index")
plt.ylabel("Bill Amount")

plt.show()
```

```python
# Bar chart comparing total_bill and tip side by side for first 10 customers
# Each customer will have two bars - one for bill, one for tip
tips.iloc[0:10, 0:2].plot(kind="bar")

plt.title("Total Bill vs Tip")
plt.xlabel("Customer Index")
plt.ylabel("Amount ($)")

plt.show()
```

```python
# Bar chart of student Math scores
df.plot(
    x="Student",
    y="Math",
    kind="bar",
    legend=False
)

plt.title("Math Scores per Student")
plt.xlabel("Student")
plt.ylabel("Marks")

plt.show()
```

```python
# Bar chart comparing all subjects per student
df.plot(
    x="Student",
    y=["Math", "Science", "English"],
    kind="bar"
)

plt.title("All Subject Scores per Student")
plt.xlabel("Student")
plt.ylabel("Marks")

plt.show()
```

## groupby() - Group and Aggregate Data

groupby() is one of the most powerful functions in Pandas. It splits the DataFrame into groups based on a column, then you apply an aggregation function to each group.

Syntax:
    df.groupby("column")["target"].aggregation()

Common aggregation functions:
- mean() : average of the group
- sum() : total of the group
- count() : number of rows in the group
- max() : highest value in the group
- min() : lowest value in the group

The result of groupby() is itself a Series or DataFrame, so we can call .plot() directly on it.

```python
# Average bill per day
# Step 1: group rows by the day column
# Step 2: select the total_bill column from each group
# Step 3: calculate the mean for each group

tips.groupby("day")["total_bill"].mean().plot(kind="bar")

plt.title("Average Bill per Day")
plt.xlabel("Day")
plt.ylabel("Average Bill")

plt.show()
```

```python
# Print the grouped result before plotting to understand what groupby returns
result = tips.groupby("day")["total_bill"].mean()
print(result)
```

```python
# Average tip per day
tips.groupby("day")["tip"].mean().plot(kind="bar")

plt.title("Average Tip per Day")
plt.xlabel("Day")
plt.ylabel("Average Tip")

plt.show()
```

```python
# Average bill for Lunch vs Dinner
tips.groupby("time")["total_bill"].mean().plot(kind="bar")

plt.title("Average Bill - Lunch vs Dinner")
plt.xlabel("Time")
plt.ylabel("Average Bill")

plt.show()
```

```python
# Average bill by gender
tips.groupby("sex")["total_bill"].mean().plot(kind="bar")

plt.title("Average Bill by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Bill")

plt.show()
```

```python
# Count of customers per day
tips.groupby("day")["total_bill"].count().plot(kind="bar")

plt.title("Number of Customers per Day")
plt.xlabel("Day")
plt.ylabel("Number of Customers")

plt.show()
```

## Interview Questions

Q1. What is Data Visualization?

Data Visualization is the process of representing data graphically using charts, graphs, and plots so that patterns, trends, and insights can be understood quickly without reading raw numbers.

---

Q2. What is a DataFrame?

A DataFrame is a two-dimensional tabular data structure in Pandas. It has labeled rows and columns, similar to an Excel spreadsheet. Each column in a DataFrame is a Series.

---

Q3. What is a Series?

A Series is a one-dimensional labeled array. When you select a single column from a DataFrame using df["column"], the result is a Series.

---

Q4. What does the plot() function do?

The plot() function creates a graph directly from Pandas data. By default it creates a line chart. You can use the kind parameter to change the chart type, for example kind="bar" for a bar chart.

---

Q5. What is iloc() and how is it different from loc()?

iloc stands for integer location. It selects rows and columns using integer position numbers. loc on the other hand uses labels or index names. For example, iloc[0:3] gives the first 3 rows by position, while loc[0:3] gives rows with index labels 0 to 3 (inclusive).

---

Q6. What is groupby()?

groupby() splits a DataFrame into groups based on the unique values in a column. You then apply an aggregation function like mean() or count() to each group. It is used to analyze data by category.

---

Q7. Why do we use Seaborn along with Matplotlib?

Seaborn provides higher-level plotting functions that produce more attractive and informative statistical graphs with less code. Matplotlib gives us fine-grained control over every element of the chart like labels and titles. They are commonly used together.

---

Q8. What is the difference between a line chart and a bar chart?

A line chart is used when data is continuous and you want to show trends or changes over time. A bar chart is used when comparing discrete, separate categories where each value is independent.
